```python
import pandas as pd
import numpy as np
from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)
```

```python
file_path = "C:/Users/Amith Nandivada/Desktop/Amith/MS/CUB/Boulder Course Material/Fall23/Machine Learning/Assignment/Data/cleaned_data.csv"

property_df = pd.read_csv(file_path)
print(property_df.info())
```

```python
cat_cols_property_df = property_df.select_dtypes(include=['object'])
cat_cols_property_df.describe()
```

```python
print(cat_cols_property_df.shape)
cat_cols_property_df = cat_cols_property_df.drop([x for x in cat_cols_property_df.columns if cat_cols_property_df[x].nunique()>6], axis=1)
print(cat_cols_property_df.shape)
```

```python
cat_cols_property_df.describe()
```

```python
for cols in cat_cols_property_df.columns:
    print("*"*100)
    print("Column Name : ", cols)
    print("\n")
    print("Unique Values :")
    print(cat_cols_property_df[cols].value_counts(normalize=True)*100)
```

```python
cat_cols_property_df['prkgType'] = cat_cols_property_df['prkgType'].str.replace(',', ' ')
cat_cols_property_df['prkgType'] = cat_cols_property_df['prkgType'].str.replace('Carport \\(Unspecified\\)', 'Carport')

# Print the updated unique values
print(cat_cols_property_df['prkgType'].value_counts(normalize=True)*100)
```

```python
cat_cols_property_df = cat_cols_property_df.apply(lambda x: x.str.upper())

transactions_list = cat_cols_property_df.values.tolist()
transactions_list
```

```python
import csv

# Full path for the output CSV file
output_csv_path = "C:/Users/Amith Nandivada/Desktop/Amith/MS/CUB/Boulder Course Material/Fall23/Machine Learning/Assignment/Data/transactional_data.csv"

# Save transaction list as CSV file
with open(output_csv_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(transactions_list)
```
